# *** seaborn.heatmap()  Most Used Parameters***


seaborn.heatmap(

    data,               # 2D dataset (e.g., DataFrame or 2D array)
    
    # Annotation
    annot=False,        # Write data values in each cell
    fmt='.2g',          # Format of annotations (e.g., '.2f', 'd')
    annot_kws=None,     # Dict to customize annotation text (e.g., {'size': 10})
    
    # Color Mapping
    cmap=None,          # Colormap (e.g., 'coolwarm', 'viridis')
    center=None,        # Value at which to center the colormap
    vmin=None,          # Minimum value of colormap
    vmax=None,          # Maximum value of colormap
    robust=False,       # Use robust quantiles for colormap limits
    
    # Grid Lines
    linewidths=0,       # Width of the lines between cells
    linecolor='white',  # Color of the lines between cells
    
    # Axis Labels
    xticklabels='auto', # Show/hide/custom x-axis labels (True/False or list)
    yticklabels='auto', # Show/hide/custom y-axis labels (True/False or list)
    
    # Masking
    mask=None,          # Boolean mask to hide certain cells
    
    # Color Bar
    cbar=True,          # Show color bar
    cbar_kws=None,      # Dict to customize color bar (e.g., {'label': 'Scale'})
    
    # Shape and Axes
    square=False,       # Force square-shaped cells
    ax=None,            # Matplotlib Axes to plot on
    
    # Performance
    vmin=None,          # Lower bound of color scale
    vmax=None,          # Upper bound of color scale
    rasterized=False,   # Rasterize the heatmap (for performance in vector graphics)
    
    # Others
    xticks_rotation='vertical' # Custom rotation (requires matplotlib)
)
