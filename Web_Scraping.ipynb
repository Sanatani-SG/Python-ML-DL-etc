```python
from bs4 import BeautifulSoup     #For html files
import requests as r

with open('Web_Scraping_Sample.html') as html_file:
  soup = BeautifulSoup(html_file, 'lxml')
for i in soup.find_all('div',class_='article'):
  print(i.text)
  print()
```

```python
#For Actual Website
s=r.get('https://www.worldometers.info/coronavirus/').text
soup = BeautifulSoup(s, 'lxml')
o=soup.find_all('div',class_='maincounter-number')
oo=soup.find_all('div',id='maincounter-wrap')
for i,j in zip(o,oo):
  h=j.find('h1')
  if h:
    print(h.text)
  print(i.text)
```
