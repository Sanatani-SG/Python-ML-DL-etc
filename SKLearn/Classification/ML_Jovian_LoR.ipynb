```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
data=pd.DataFrame(pd.read_csv('weatherAUS.csv'))
data.dropna(subset=['RainTomorrow'], inplace=True)
```

```python
data
```

```python
data[['RainTomorrow','RainToday']]=data[['RainTomorrow','RainToday']].replace({'No':0,'Yes':1})
data
```

```python
i=SimpleImputer(strategy='mean')
e=OneHotEncoder()
s=MinMaxScaler()
year=pd.to_datetime(data['Date']).dt.year
X=data[year<2014]
Y=data[year==2014]
Z=data[year>2014]
X,Y,Z=X.drop('Date',axis=1),Y.drop('Date',axis=1),Z.drop('Date',axis=1)
```

```python
X_inp=X.drop('RainTomorrow',axis=1)
X_tar=X['RainTomorrow']
Y_inp=Y.drop('RainTomorrow',axis=1)
Y_tar=Y['RainTomorrow']
Z_inp=Z.drop('RainTomorrow',axis=1)
Z_tar=Z['RainTomorrow']
```

```python
obj=list(X_inp.select_dtypes(include='object').columns)
X_inp[obj].columns
```

```python
# Fit encoder on all data for consistent categories
Xe=e.fit_transform(X_inp[obj])

# # Transform X_inp[obj] and create DataFrame with correct index
Xe = e.transform(X_inp[obj])
encoded_cols = e.get_feature_names_out(obj)
encoded_df = pd.DataFrame(Xe.toarray() if hasattr(Xe, "toarray") else Xe, columns=encoded_cols, index=X_inp.index)

# # Drop original categorical columns and concatenate encoded columns
X_inp = pd.concat([X_inp.drop(obj, axis=1), encoded_df], axis=1)
```
