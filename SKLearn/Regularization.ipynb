```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.DataFrame(pd.read_csv('melb_data.csv'))
```

```python
data.info()
```

```python
data.describe(include='all')
```

```python
a=data.columns
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount','Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
dataset = data[cols_to_use]
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
dataset['Car']=dataset['Car'].fillna(0)
```

```python
dataset.isna().sum()
```

```python
dataset['BuildingArea']=dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='most_frequent')
dataset['CouncilArea'] = imputer.fit_transform(dataset[['CouncilArea']]).flatten()
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
x=dataset.drop('Price',axis=1)
y=dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
Tx,tx,Ty,ty=train_test_split(x,y,test_size=0.3,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(Tx,Ty)
```

```python
reg.score(Tx,Ty)
```

```python
reg.score(tx,ty)
```

```python
from sklearn.linear_model import Lasso,Ridge
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
for i in [0.001, 0.01, 0.1, 1, 10, 50]:
    model = make_pipeline(StandardScaler(), Lasso(alpha=i, max_iter=10000, tol=0.0001))
    model.fit(Tx, Ty)
    print(f"alpha={i}, train score={model.score(Tx, Ty)},test score={model.score(tx, ty)}")
```

```python
for i in [0.001, 0.01, 0.1, 1, 10, 50]:
    model = make_pipeline(StandardScaler(), Ridge(alpha=i, max_iter=10000, tol=0.0001))
    model.fit(Tx, Ty)
    print(f"alpha={i}, train score={model.score(Tx, Ty)},test score={model.score(tx, ty)}")
```
