```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
data=pd.DataFrame(pd.read_csv('data.csv'))
```

```python
val,test=train_test_split(data,test_size=0.2)
train,val=train_test_split(val,test_size=0.25)
```

```python
train
```

```python
ty=train['diagnosis']
train=train.drop(['diagnosis','id','Unnamed: 32'],axis=1)
vy=val['diagnosis']
val=val.drop(['diagnosis','id','Unnamed: 32'],axis=1)
Ty=test['diagnosis']
test=test.drop(['diagnosis','id','Unnamed: 32'],axis=1)
```

```python
model1=DecisionTreeClassifier(max_depth=10,max_leaf_nodes=5)
model1.fit(train,ty)
```

```python
accuracy_score(ty,model1.predict(train))
```

```python
accuracy_score(vy,model1.predict(val))
```

```python
accuracy_score(Ty,model1.predict(test))
```

```python
model2=RandomForestClassifier(n_estimators=100,max_depth=100,max_leaf_nodes=50)
model2.fit(train,ty)
```

```python
accuracy_score(ty,model2.predict(train))
```

```python
accuracy_score(vy,model2.predict(val))
```

```python
accuracy_score(Ty,model2.predict(test))
```

```python
model3=LogisticRegression(solver='liblinear',max_iter=100)
model3.fit(train,ty)
```

```python
accuracy_score(ty,model3.predict(train))
```

```python
accuracy_score(vy,model3.predict(val))
```

```python
accuracy_score(Ty,model3.predict(test))
```
