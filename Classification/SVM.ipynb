```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
dir(iris)
```

```python
data=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
data
```

```python
iris.target_names
```

```python
data['Target']=iris.target
```

```python
data['Flower']=data.Target.apply(lambda x:iris.target_names[x])
```

```python
data
```

```python
d1=data[data.Target==1]
d0=data[data.Target==0]
d2=data[data.Target==2]
```

```python
d1
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(d1['sepal length (cm)'],d1['sepal width (cm)'],c='red')
# plt.scatter(d2['sepal length (cm)'],d2['sepal width (cm)'],c='blue',marker='*')
plt.scatter(d0['sepal length (cm)'],d0['sepal width (cm)'],c='orange',marker='+')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.drop(['Target',	'Flower'],axis=1)
y=data['Target']
```

```python
tx,Tx,ty,Ty=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
model.fit(tx,ty)
```

```python
model.score(Tx,Ty)
```
