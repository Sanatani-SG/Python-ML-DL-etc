```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
d=load_digits()
```

```python
d.data
```

```python
d.data[0]
```

```python
plt.matshow(d.images[0])
```

```python
x,X,y,Y=train_test_split(d.data,d.target,test_size=0.2)
```

```python
model=LogisticRegression()
model.fit(x,y)
```

```python
model.score(X,Y)
```

```python
from joblib import dump,load
```

```python
dump(model,'model.joblib')
```

```python
from tensorflow.keras.datasets import mnist
```

```python
(x,y),(X,Y)=mnist.load_data()
```

```python
model=LogisticRegression(max_iter=250) # Increased max_iter to avoid convergence warning
model.fit(x.reshape(-1, 28*28), y) # Reshape x to 2D
```

```python
X=X.reshape(-1, 28*28)
```

```python
model.predict(X)
```

```python
model.score(X,Y)
```
